```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('1.jpg', 0)

# show image as black and white
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
hist = cv2.calcHist([img],[0], None, [256], [0,256])
plt.plot(hist)
plt.show()
```

```python
img1 = cv2.imread('1.jpg',0)
hist2 = cv2.calcHist([img1], [0], None, [256], [0,256])
plt.plot(hist2)
plt.show()
```

```python
plt.subplot(121), plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.subplot(122), plt.plot(hist2)
```

```python
hist3, bins = np.histogram(img1.ravel(), 256, [0,256])
plt.plot(hist3)
plt.show()
```

```python
img3 = cv2.imread('1.jpg',1)
color = ('b','g','r')
for i,col in enumerate(color):
    hist4 = cv2.calcHist([img3], [i], None, [256],[0,256])
    plt.plot(hist4,color = col)
    plt.xlim(0,255)
```

```python
b, g, r = cv2.split(img3)

hist1, bins = np.histogram(b.ravel(), 256, [0,256])
hist2, bins = np.histogram(g.ravel(), 256, [0,256])
hist3, bins = np.histogram(r.ravel(), 256, [0,256])

plt.plot(hist1)
plt.plot(hist2)
plt.plot(hist3)

plt.xlim([0,256])
plt.show()
```

```python
b, g, r = cv2.split(img3)

plt.hist(b.ravel(), 256, [0,256])
plt.hist(g.ravel(), 256, [0,256])
plt.hist(r.ravel(), 256, [0,256])

plt.xlim([0,256])
plt.show()
```

```python
mask = np.zeros(img3.shape[:2], np.uint8)

mask[50:300,250:400] = 255

plt.imshow(cv2.cvtColor(mask,cv2.COLOR_BGR2RGB))
```

```python
maskedimg = cv2.bitwise_and(img3,img3, mask = mask)
plt.imshow(cv2.cvtColor(maskedimg,cv2.COLOR_BGR2RGB))
```

```python
hist2 = cv2.calcHist([maskedimg], [0], None, [256], [0,256])
plt.plot(hist2)
plt.show()
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    hist4 = cv2.calcHist([img3], [i], None, [256],[0,256])
    plt.plot(hist4,color = col)
    plt.xlim(0,255)
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    hist4 = cv2.calcHist([maskedimg], [i], None, [256],[0,256])
    plt.plot(hist4,color = col)
    plt.xlim(0,255)
```
